```python
import pickle
import pandas as pd
import numpy as np
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt
```

### Doc2vec & RF

```python
features_train = pickle.load(open('../doc2vec_results/features_train_doc2vec.pkl', 'rb'))
model = pickle.load(open('../doc2vec_results/model_doc2vec_RF.pkl', 'rb'))
predictions = model.predict(features_train)
```

```python
n, bins, patches = plt.hist(predictions, 100)
plt.xlabel('Prediction')
plt.ylabel('Number of predictions')
plt.title('Distribution of preditions for Doc2vec and RF')
# plt.axis([40, 160, 0, 0.03])
plt.grid(True)
plt.show()
```
